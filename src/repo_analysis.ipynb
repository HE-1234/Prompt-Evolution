```python
import pandas as pd
from git import Repo
import os
import traceback
import ast
import re
```

```python
REPO_PATH = "filtered_data/stars_50_activity_6.csv"
base_url = "https://github.com/"
```

```python
base_path = 'repositories/'
os.makedirs(base_path, exist_ok=True)
```

```python
def clone_repo(repo_identifier, target_path):
    repo_url = f"{base_url}{repo_identifier}.git"
    # target_path = os.path.join(base_path, repo_identifier.split('/')[-1])
    if not os.path.exists(target_path):  # Only clone if the directory doesn't exist
        try:
            print(f"Cloning {repo_url} into {target_path}")
            Repo.clone_from(repo_url, target_path)
            print("Successfully cloned.")
        except Exception as e:
            print(f"Failed to clone {repo_url}. Error: {e}")
    else:
        print(f"Repository already exists at {target_path}")

def count_commits(repo_path, file_path):
    try:
        repo = Repo(repo_path)
        repo.iter_commits(paths=file_path)
        commits_count = len(list(repo.iter_commits(paths=file_path)))
        return commits_count
    except Exception as e:
        print(f"Error counting commits for {file_path} in {repo_path}:")
        print(e)
        traceback.print_exc()
        return 0

def count_commits_with_prompt(repo_path, file_path):
    try:
        repo = Repo(repo_path)
        prompts_in_commit_count = 0
        
        for commit in repo.iter_commits(paths=file_path):
            if 'prompt' in commit.message.lower():
                prompts_in_commit_count += 1
                
        return prompts_in_commit_count
    except Exception as e:
        print(f"Error processing {file_path} in {repo_path}: {e}")
        return 0
```

```python
df = pd.read_csv(REPO_PATH)
print(len(df['repo_name'].unique()))
```

```python
print(len(df))
```

```python
for index, repo in enumerate(df['repo_name'].unique()):
    print(f"{index}/{len(df['repo_name'].unique())}")
    repo_path = os.path.join(base_path, repo)  # Assuming repo_name is a URL or user/repo format
    if not os.path.exists(repo_path):
        clone_repo(repo, repo_path)
```

```python
df['commit_count'] = 0
# df['prompt_commit_count'] = 0

for index, row in df.iterrows():
    print(f"Counting Commits for {row['repo_name']} {index}/{len(df)}")
    repo_path = os.path.join(base_path, row['repo_name'])
    file_path = row['file_name'].replace('~', '/')  # Convert to standard file path
    df.at[index, 'commit_count'] = count_commits(repo_path, file_path)
    # df.at[index, 'prompt_commit_count'] = count_commits_with_prompt(repo_path, file_path)
```

```python
def parse_quoted_strings(input_string):
    # Strip the surrounding square brackets
    if input_string.startswith('[') and input_string.endswith(']'):
        input_string = input_string[1:-1].strip()
    
    result = []
    i = 0
    while i < len(input_string):
        # Skip leading spaces and newlines
        if input_string[i] in "\"'":
            current_quote = input_string[i]
            i += 1
            start_index = i
            # Handle nested quotes by counting
            
            while i < len(input_string):
                if input_string[i] == current_quote:
                    result.append(input_string[start_index:i])
                    i += 1
                    break
                i += 1
        else:
            i += 1
            continue
    return result

def get_max_prompt_length(prompts):
    if not prompts:
        return 0
    return max(len(prompt) for prompt in prompts)

def get_max_prompt_words(prompts):
    if not prompts:
        return 0
    # Count the number of words in each prompt by splitting by whitespace and return the max count
    return max(len(prompt.split()) for prompt in prompts)
```

```python
for index, row in df.iterrows():
    print(f"parsing row {index}")
    df.at[index, 'prompts'] = parse_quoted_strings(row['prompts'])
    
```
